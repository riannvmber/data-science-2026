```python
#Load Dataset Titanic
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

df.head()
```

```python

#Kode Q1
print("Shape Dataset:")
print(df.shape)

print("\nDaftar Kolom:")
print(df.columns.tolist())
```

```python
missing = df.isnull().sum()

persentase = (missing / len(df)) * 100

hasil_missing = pd.DataFrame({
    'Jumlah Missing': missing,
    'Persentase (%)': persentase
})

hasil_missing[hasil_missing['Jumlah Missing'] > 0]
```

```python
persen_selamat = df["Survived"].mean() * 100

print(f"Persentase penumpang yang selamat: {persen_selamat:.2f}%")
```

```python
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
jumlah = wanita_kelas1.shape[0]
rata_usia = wanita_kelas1['Age'].mean()

print("jumlah wanita kelas 1:", jumlah)
print("Rata-rata usia:", round(rata_usia, 2))
```

```python
hasil = pd.DataFrame({
    "Persentase Selamat (%)":
    df.groupby("Pclass")["Survived"].mean() * 100
})

hasil
```

```python
hasil_gender = pd.DataFrame({
    "Persentase Selamat (%)":
    df.groupby("Sex")["Survived"].mean() * 100 })

hasil_gender
```
